```python
#Change the country to highlight value to your selected country

def plot_data(df, country_to_highlight):
    # Ensure the necessary columns are present
    if 'Happiness Score' in df.columns and 'Economy' in df.columns and 'Country' in df.columns:
        plt.figure(figsize=(10, 6))
        
        # Plot all data points
        plt.scatter(df['Economy'], df['Happiness Score'], alpha=0.5, label='Other countries')
        
        # Highlight the specified country
        if country_to_highlight in df['Country'].values:
            country_data = df[df['Country'] == country_to_highlight]
            plt.scatter(country_data['Economy'], country_data['Happiness Score'], color='red', label=country_to_highlight, edgecolors='black', s=100)
        
        # Add titles and labels
        plt.title('Happiness Score vs Economy')
        plt.xlabel('Economy')
        plt.ylabel('Happiness Score')
        plt.legend()
        plt.show()
    else:
        print("The dataset does not contain the required columns.")

# Call the function to plot the data and highlight a specific country
country_to_highlight = 'Nigeria'  # Replace with the country you want to highlight
plot_data(df, country_to_highlight)
```
